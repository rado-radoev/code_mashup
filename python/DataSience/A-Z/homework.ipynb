```python
def myplot(data, playerlist=Players):
    
    colors = {"KobeBryant":"Black","JoeJohnson":"Red","LeBronJames":"Blue","CarmeloAnthony":"Yellow","DwightHoward":"Magenta","ChrisBosh":"Grey","ChrisPaul":"Brown","KevinDurant":"Black","DerrickRose":"Green","DwayneWade":"Purple"}
    marker= {"KobeBryant":".","JoeJohnson":"s","LeBronJames":"D","CarmeloAnthony":"X","DwightHoward":"s","ChrisBosh":"8","ChrisPaul":"p","KevinDurant":"P","DerrickRose":"h","DwayneWade":"H"}
    
    for name in playerlist:
        plt.plot(data[Pdict[name]], c=colors[name], ls='--', marker=marker[name], ms='7', label=name)
    
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0, 10)), Seasons, rotation='vertical')
    plt.show()
 
```
